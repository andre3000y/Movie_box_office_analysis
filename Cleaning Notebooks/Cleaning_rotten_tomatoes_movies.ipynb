```python
import sqlite3
import pandas as pd
import seaborn as sns

conn = sqlite3.connect('../data/movies.db')

df = pd.read_sql('''
select * from rotten_tomatoes_movies
''', conn)
df.head()
```

```python
df.info()
```

```python
df.drop(['critics_consensus','movie_info', 'original_release_date','streaming_release_date'], axis = 1, inplace = True)
df.info()
```

```python
df.loc[df['genres'].isna()]
```

```python
def listize(stringlist):
    if stringlist == None:
        return []
    list_version = stringlist.split(',')
    return list_version

def columnlister(listcols):
    for col in listcols:
        df[col] = df[col].map(listize)
```

```python
columnlister(['genres','directors','authors','actors'])
```

```python
df.info()
```

```python
df['production_company'].fillna('No Company', inplace = True)
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df.info()
```
